```python
from ColabTurtle.Turtle import *
import colorsys

initializeTurtle()
speed(13)
width(2)

num_petalos = 16

def dibujar_petalo():
    for _ in range(2):
        for _ in range(90):
            forward(2)
            left(1)
        left(90)

for i in range(num_petalos):
    hsv_color = colorsys.hsv_to_rgb(i / num_petalos, 1.0, 1.0)
    rgb_val = tuple(int(x * 255) for x in hsv_color)

    color(rgb_val)

    dibujar_petalo()

    left(360 / num_petalos)

hideturtle()
```
