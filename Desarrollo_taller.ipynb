## 🧠 1. ¿Qué es un diccionario?

Un **diccionario** es una estructura de datos que guarda información en pares:

- **clave**: el identificador único
- **valor**: el dato asociado a esa clave

### Ejemplo de uso

Imagina que tu "inventario" de tu inventario se genera así:

- `"Pala": 15`
- `"Exp": 10000`
- `"Kinket": 12000`

Esto significa que, posees tres objetos:

- la clave `"Pala"` tiene como valor `15`
- la clave `"Exp"` tiene como valor `10000`
- la clave `"Kinket"` tiene como valor `12000`

Los diccionarios son útiles cuando quieres buscar datos rápidamente usando un nombre o identificador.

## 🐍 2. Diccionarios en Python

En Python, los diccionarios se escriben con llaves `{}`.

### Sintaxis general

```python
diccionario = {
    "clave1": valor1,
    "clave2": valor2
}
```

````python
# Inventario del minero
inventario = {
    "Pala": 15,
    "Exp": 10000,
    "Kinket": 12000
}

print(inventario)
````

### Operaciones básicas en Python

````python
inventario = {
    "Pala": 15,
    "Exp": 10000,
    "Kinket": 12000
}

# 1. Acceder a un valor
print("Daño de la pala:", inventario["Pala"])

# 2. Agregar un nuevo elemento
inventario["pocion"] = 3
print("Después de agregar poción:", inventario)

# 3. Modificar un valor existente
inventario["Kinket"] = 200
print("Después de modificar oro:", inventario)

# 4. Eliminar un elemento
del inventario["Exp"]
print("Después de eliminar escudo:", inventario)
````

## ☕ 3. Diccionarios en Java con `HashMap`

En Java, la estructura equivalente más común es `HashMap`.

### Sintaxis general

```java
import java.util.HashMap;

HashMap<String, Integer> inventario = new HashMap<>();
inventario.put("Pala", 15);
inventario.put("Exp", 10000);
inventario.put("Kinket", 12000);
```

### Operaciones básicas en Java

```java
import java.util.HashMap;

HashMap<String, Integer> inventario = new HashMap<>();

inventario.put("Pala", 15);
inventario.put("Exp", 10000);
inventario.put("Kinket", 12000);

// Acceder
System.out.println(inventario.get("Pala"));

// Agregar
inventario.put("pocion", 3);

// Modificar
inventario.put("Kinket", 200);

// Eliminar
inventario.remove("Exp");
```

### Idea clave

- En **Python** usamos `dict`
- En **Java** usamos con frecuencia `HashMap`
- En ambos casos la lógica es la misma: **clave → valor**

## ⚔️ 4. Ejercicio explicado 1 — Nivel muy sencillo


La archivista del reino guardó el poder de tres reliquias:

- espada: 20
- arco: 12
- bastón: 18

Se desea mostrar el poder de la espada.

---

### ✅ Solución en Python

````python
reliquias = {
    "espada": 20,
    "arco": 12,
    "bastón": 18
}

print("El poder de la espada es:", reliquias["espada"])
````

### 🔍 Explicación paso a paso

1. Se crea un diccionario llamado `reliquias`.
2. Cada reliquia tiene una clave:
   - `"espada"`
   - `"arco"`
   - `"bastón"`
3. Para consultar un valor, se usa la clave entre corchetes:
   `reliquias["espada"]`
4. El resultado es `20`.

---

### ✅ Solución equivalente en Java

```java
import java.util.HashMap;

HashMap<String, Integer> reliquias = new HashMap<>();

reliquias.put("espada", 20);
reliquias.put("arco", 12);
reliquias.put("baston", 18);

System.out.println("El poder de la espada es: " + reliquias.get("espada"));
```

## 🟡 5. Ejercicio explicado 2 — Complejidad media


Un guardián registra cuántas veces aparecen ciertos monstruos durante una expedición.

Lista observada:

`["slime", "dragon", "slime", "lobo", "dragon", "slime"]`

Se desea construir un diccionario que cuente cuántas veces aparece cada monstruo.

---

### ✅ Solución en Python

````python
monstruos = ["slime", "dragon", "slime", "lobo", "dragon", "slime"]
conteo = {}

for monstruo in monstruos:
    if monstruo in conteo:
        conteo[monstruo] += 1
    else:
        conteo[monstruo] = 1

print("Conteo final:", conteo)
````

### 🔍 Explicación paso a paso

1. Se parte de una lista de monstruos repetidos.
2. Se crea un diccionario vacío llamado `conteo`.
3. Se recorre la lista con un ciclo `for`.
4. En cada vuelta:
   - si el monstruo ya está en el diccionario, se suma 1
   - si no está, se crea con valor 1
5. Al final, el diccionario contiene la frecuencia de aparición.

### Resultado esperado

```python
{'slime': 3, 'dragon': 2, 'lobo': 1}
```

---

### ✅ Solución equivalente en Java

```java
import java.util.HashMap;

HashMap<String, Integer> conteo = new HashMap<>();
String[] monstruos = {"slime", "dragon", "slime", "lobo", "dragon", "slime"};

for (String monstruo : monstruos) {
    if (conteo.containsKey(monstruo)) {
        conteo.put(monstruo, conteo.get(monstruo) + 1);
    } else {
        conteo.put(monstruo, 1);
    }
}

System.out.println(conteo);
```

## 🧩 6. Resumen rápido: Python vs Java

| Aspecto | Python | Java |
|---|---|---|
| Estructura | `dict` | `HashMap` |
| Crear | `{}` | `new HashMap<>()` |
| Insertar | `dic["x"] = 5` | `map.put("x", 5)` |
| Consultar | `dic["x"]` | `map.get("x")` |
| Verificar existencia | `"x" in dic` | `map.containsKey("x")` |
| Eliminar | `del dic["x"]` | `map.remove("x")` |

### Idea central

Aunque cambie la sintaxis, en ambos lenguajes se trabaja con la misma lógica:

**usar una clave para encontrar un valor**
