## **Tabla de multiplicar hasta el 10**

```python
numero = int(input("Ingrese un número: "))

for i in range(1, 11):
    print(numero, "x", i, "=", numero * i)
```

## **Suma de numeros naturales hasta n**

```python
n = int(input("Ingrese un número n: "))
suma = 0

for i in range(1, n + 1):
    suma += i

print("La suma es:", suma)
```

## **Factorial de un numero**

```python
n = int(input("Ingrese un número: "))

import math

resultado = math.factorial(n)

print("El factorial es:", resultado)
```

## **Serie Fibonacci**

```python
n = int(input("¿Cuántos números de Fibonacci desea ver?: "))

a = 0
b = 1
contador = 0

while contador < n:
    print(a)
    siguiente = a + b
    a = b
    b = siguiente
    contador += 1
```

## **Adivinar un color**

```python
import random

colores = []

print("Ingrese 10 colores:")
for i in range(10):
    color = input("Color " + str(i+1) + ": ")
    colores.append(color)

secreto = random.choice(colores)
intentos = 0

while intentos < 10:
    intento = input("Adivine el color: ")
    intentos += 1

    if intento == secreto:
        print("¡Correcto!")
        break
    else:
        print("Incorrecto, intenta de nuevo.")

if intentos == 10 and intento != secreto:
    print("Se acabaron los intentos. El color era:", secreto)
```

## **Suma de los digitos de un numero**

```python
numero = int(input("Ingrese un número: "))
suma = 0

while numero > 0:
    digito = numero % 10
    suma += digito
    numero //= 10

print("La suma de los dígitos es:", suma)
```

## **Determinar si un numero es primo**

```python
numero = int(input("Ingrese un número: "))

if numero <= 1:
    print("No es primo")
else:
    es_primo = True
    for i in range(2, numero):
        if numero % i == 0:
            es_primo = False
            break

    if es_primo:
        print("Es primo")
    else:
        print("No es primo")
```
